## Let's take a look at video game sales

### NA Sales = North America Sales (in millions)
### EU Sales = Europe Sales (in millions)
### JP Sales = Japan Sales (in millions)

```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load data
vgdata = pd.read_csv("videogamesales.csv")
```

```python
vgdata.head()
```

### Total Sales per Genre

```python
#Gspg = Global Sales per Genre
gspg = vgdata.groupby(['Genre'])['Global_Sales'].sum()
gspg  = gspg.reset_index().sort_values('Global_Sales', ascending = False)
```

```python
gspg.head()
```

```python
#Calculate totals and proportions per region
total_NA_sales = vgdata['NA_Sales'].sum()
total_EU_sales = vgdata['EU_Sales'].sum()
total_JP_sales = vgdata['JP_Sales'].sum()
total_Other_sales = vgdata['Other_Sales'].sum()
total_Global_sales = vgdata['Global_Sales'].sum()

NA_prop = round((total_NA_sales / total_Global_sales), 4)
EU_prop = round((total_EU_sales / total_Global_sales), 4)
JP_prop = round((total_JP_sales / total_Global_sales), 4)
Oth_prop = round((total_Other_sales / total_Global_sales), 4)
```

```python
#Visualize the data
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'NA','EU', 'JP','Other'
values = [NA_prop, EU_prop, JP_prop, Oth_prop]

fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
